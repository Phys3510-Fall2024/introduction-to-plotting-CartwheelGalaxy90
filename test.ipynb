```python
import matplotlib.pyplot as plt

# Example data
x = range(10)
y1 = [xi**2 for xi in x]
y2 = [xi**1.5 for xi in x]

# Plotting multiple lines
plt.plot(x, y1, label='x squared')
plt.plot(x, y2, label='x to the power of 1.5')
plt.legend()
plt.title('Multi-Line Graph')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```
