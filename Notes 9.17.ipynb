## Plotting with `matplotlib`

```python
# import numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define the potential energy
def potential_energy(x, k=1.0):
    return 0.5 * k * x**2 # k is the spring constant

# test function for the potential energy
potential_energy(2,10)
```

```python
# define mesh for plotting
x = np.linspace(-5, 5, 100) # np.linspace(x_min, x_max, npts)
y = potential_energy(x)

# testing the output
# print(np.transpose([x, y]))
```

```python
# plot the function

plt.figure(figsize=(4,3))
plt.plot(x, y, color="black", linewidth=0.5)

# add axes label
plt.xlabel("displacement (arbitrary units)")
plt.ylabel("potential energy (arbitrary units)")

# change the limits
plt.xlim([-4,4])
plt.ylim([-1,10])

plt.show() # combine all elements of the plot
```

## Contour Plots

```python
# import numpy and matplotlib
import numpy as np
import matplotlib.pyplot as plt
```

```python
# define the potential energy
def potential_energy_2D(x, y, kx=1.0, ky=1.0):
    return 0.5 * (kx * x**2 + ky * y**2) # kx, ky are spring constants

# test function for the potential energy
potential_energy_2D(2,1)
```

```python
# define mesh for plotting
x = np.linspace(-5, 5, 100) # np.linspace(x_min, x_max, npts)
y = np.linspace(-5, 5, 100) # np.linspace(y_min, y_max, npts)

# combine 1D meshes into meshgrid
X, Y = np.meshgrid(x,y)

# define Z values
Z = potential_energy_2D(X,Y)

# testing the mesh
# print(X)
# print()
# print(Y)
# print()
# print(Z)
```

```python
# construct contour plot
plt.figure(figsize=(4,3)) # figure size in inches
plt.contourf(X, Y, Z, levels=10)

# add a colorbar legend
plt.colorbar(label="potential energy (arbitrary units)")

# add axes label
plt.xlabel("x-displacement (arbitrary units)")
plt.ylabel("y-potential energy (arbitrary units)")

# change the limits
plt.xlim([-5,5])
plt.ylim([-5,5])

plt.show() # combine all elements of the plot
```
