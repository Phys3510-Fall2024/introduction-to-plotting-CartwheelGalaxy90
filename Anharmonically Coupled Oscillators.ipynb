```python
# Import the enecessary packages
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
```

```python
# Define the potential energy
def potential_energy(Q1, Q2, K1=1, K2=1, C=1):
    return 0.5*K1*Q1**2 + 0.5*K2*Q2**2 + C*Q1*Q2**2  # K1 and K2 are spring constants

# Test the function
# potential_energy(1,2)
```

```python
# Define the meshes and meshgrid
q1 = np.linspace(-1,1,10)
q2 = q1
Q1, Q2 = np.meshgrid(q1, q2)

# potential_energy(Q1, Q2)
```

```python
# Plot with Q2=0
#plt.plot(potential_energy(Q1, Q2=0))
#plt.plot(potential_energy(Q1, Q2=-2))
#plt.plot(potential_energy(Q1, Q2=-3))
#plt.plot(potential_energy(Q1, Q2=2))
#plt.plot(potential_energy(Q1, Q2=3))
plt.plot(Q1, Q2, potential_energy(Q1, Q2, K1=1, K2=1, C=1))

plt.xlabel('$Q_1$')
plt.ylabel('$Q_2$')

plt.show()
```

```python
# plot contours with matplotlib contourf
plt.figure(figsize=(6,5)) # Sizing plot

contourf = plt.contourf(Q1, Q2, potential_energy(Q1, Q2, K1=0.5, K2=1, C=.5), levels=25)

cbar = plt.colorbar(contourf)  # Add a color bar

plt.xlim([-1,1])
plt.ylim([-1,1])
plt.xlabel('$Q_1$')
plt.ylabel('$Q_2$')

plt.show()
```
