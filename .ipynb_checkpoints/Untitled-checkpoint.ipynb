```python
# Import the enecessary packages
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the potential energy
def potential_energy(Q1, Q2, K1=1, K2=1, C=1):
    return 0.5*K1*Q1**2 + 0.5*K2*Q2**2 + C*Q1*Q2**2  # K1 and K2 are spring constants

# Define the meshes and meshgrid
q1 = np.linspace(-1,1,2)
q2 = q1
Q1, Q2 = np.meshgrid(q1, q2)
```

```python
# Create a figure with multiple subplots (2 rows, 1 column)
plt.figure(figsize=(8, 3))

# First subplot: Potential Energy (Parabola)
plt.subplot(1, 2, 1)  # 2 rows, 1 column, first plot
plt.plot(potential_energy(Q1, Q2=0), color='purple', linestyle='-', linewidth=1, label='Potential Energy $Q_2$=0')
plt.plot(potential_energy(Q1, Q2=-2), color='red', linestyle='--', linewidth=1, label='Potential Energy $Q_2$=-2')
plt.xlabel('$Q_1$', fontsize=12)
plt.ylabel('$Q_2$', fontsize=12)
plt.title('Potential Energy of Anharmonic Oscillator', fontsize=12)
plt.grid(True, color='gray', linestyle='--', linewidth=0.5)
plt.legend(fontsize=12)
#plt.xlim([-1, 3])
#plt.ylim([-1, 10])
#plt.xticks(fontsize=12)
#plt.yticks(fontsize=12)

# Define the meshes and meshgrid
q1 = np.linspace(-1,1,3)
q2 = q1
Q1, Q2 = np.meshgrid(q1, q2)
# Second subplot: Force (Linear)
plt.subplot(1, 2, 2)  # 2 rows, 1 column, second plot
plt.plot(Q1, Q2, potential_energy(Q1, Q2, K1=1, K2=1, C=1), color='blue', linestyle='-', linewidth=1, label='Poential Enegry')
plt.xlabel('$Q_1$', fontsize=12)
plt.ylabel('$Q_2$', fontsize=12)
plt.title('Potential Energy', fontsize=12)
plt.grid(True, color='gray', linestyle='--', linewidth=0.5)
#plt.legend(fontsize=12)
#plt.xlim([-1, 1])
#plt.ylim([-1, 1])
#plt.xticks(fontsize=12)
#plt.yticks(fontsize=12)

# Adjust layout to prevent overlap
plt.tight_layout()

# Show the plot
plt.show()
```
